```python
#here we use to retrive the experience level count by the work year

subset = data[['work_year', 'experience_level']] #This creates a new DataFrame called subset containing only these two columns.


# It essentially creates groups based on each unique combination of values in these two columns.

work_experience_counts = subset.groupby(['work_year', 'experience_level']).size()  #This calculates the size of each group, i.e., the number of occurrences of each unique combination of values.

print(work_experience_counts)
```

```python
#retriving the employment type and there count by the work year

subset = data[['work_year', 'employment_type']]  #This creates a new DataFrame called subset containing only these two columns.

#this part group the dataframe subset by unique combination of valuesin these 2 column.
work_employment_counts = subset.groupby(['work_year', 'employment_type']).size()  #this is use calculate the size of each group.


print(work_employment_counts)
```

```python
# Grouping data by job title and calculating maximum salary for each job title

maxsalary_by_job_title = data.groupby('job_title')['salary'].max() # is a Series object that contains the max salary for each unique job title in the dataset.

# Sorting the maximum salaries in descending order and selecting the top 5

print('top_jobtitles_with_max_salary',maxsalary_by_job_title.sort_values(ascending=False).head(10)) #sorts the maximum salaries in descending order, so the highest maximum salary comes first.
```

```python
# Grouping data by job title and calculating minimum salary for each job title

minsalary_by_job_title = data.groupby('job_title')['salary'].min() #is a Series object that contains the minimum salary for each unique job title in the dataset.

# Sorting the maximum salaries in descending order and selecting the top 5

print('top_jobtitles_with_max_salary',minsalary_by_job_title.sort_values(ascending=True).head(10)) #sorts the maximum salaries in descending order, so the highest maximum salary comes first.
```

# company size analysis with pie chart

```python

company_size_counts = data['company_size'].value_counts() #count the occurence of unique values in the series.
plt.figure(figsize=(8, 8))      #This line initializes a new figure for plotting with matplotlib and sets its size to be 8x8 inches.
plt.pie(company_size_counts, labels=company_size_counts.index,autopct='%1.1f%%')  #Display the percentage value with one digit after the decimal point.
plt.title('Company Size Distribution')
plt.show()
```

# job title analysis/top 5 job titles

```python
#plt.figure(figsize=(10, 6))

#creat veriable job_title_counts which stores the counts of unique values in the 'job_title' column of the DataFrame
# data.value_counts() is a pandas function that counts the occurrences of unique values in a Series.
job_title_counts = data['job_title'].value_counts()

top_5_job_titles = job_title_counts.head(5)  #This line selects the top 5 most frequent job titles from the job_title_counts variable. 

color=['g','r','b','y']
sns.barplot(x=top_5_job_titles.index, y=top_5_job_titles.values,palette=color)
plt.title('Top 5 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=45,color='r',fontsize=10)
plt.show()
```

# Top 5 Highest Salaries by Job Title

```python
plt.figure(figsize=(10, 6))


#This line sorts the DataFrame data by the 'salary' column in descending order 
data_sorted = data.sort_values(by='salary', ascending=False)

colo =['g','r','b','y']

#This line selects the top 5 rows with the highest salaries from the data_sorted DataFrame.
top_5_salaries = data_sorted.head(5)

sns.barplot(x='job_title', y='salary', data=top_5_salaries,palette=colo)
plt.title('Top 5 Highest Salaries by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary')
#plt.xticks(rotation=45)
plt.show()
```

```python
#salary distribution by experience level of the employee.

#plt.figure(figsize=(12, 8))
sns.barplot(x='experience_level', y='salary_in_usd',data=data,palette='Accent',saturation=100,ci=None) 
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.xticks(rotation=45,color='b')
plt.show()
```

```python
#salary in USD over a time by work year

plt.figure(figsize=(12, 6))
sns.lineplot(x='work_year', y='salary_in_usd', data=data, ci = None,marker='o')
plt.title('Trend of Average Salary Over Time')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.show()
```

```python
#here we analyse the salary distribution by their experience level and also by employment type

#plt.figure(figsize=(12, 8))

c_palette = ['g','y','b','r']
sns.barplot(x='experience_level', y='salary_in_usd',data=data,palette=c_palette,ci=100,saturation=70,hue='employment_type',errcolor='g',errwidth=2)

# hue is use to differentiate the column using another column

plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
#salary distribution by employment type.

sns.violinplot(x='employment_type', y='salary_in_usd', data=data, palette='viridis')
plt.title('Salary Distribution by employment type')
plt.xlabel('employment_type')
plt.ylabel('Salary')
plt.xticks(rotation=45,color='r',fontsize=15)
plt.show()
```

```python
#salary distribution by company size.

plt.figure(figsize=(12, 8))
sns.barplot(x='company_size', y='salary_in_usd', data=data, palette='viridis')
plt.title('Salary Distribution by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary')
plt.xticks(rotation=20,fontsize=25,color='g')
plt.show()
```

```python
# ploting the Histogram for salaries in USD

plt.figure(figsize=(10, 6))
sns.histplot(data['salary_in_usd'], kde=True, bins=30)  # kde is an argument for line on estimated value
plt.title('Distribution of Salaries in USD')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
#distribution of remote ration.

plt.figure(figsize=(10, 6))
sns.histplot(data['remote_ratio'], color='green')
plt.title('Distribution of Remote Ratio')
plt.xlabel('Remote Ratio')
plt.ylabel('Frequency')
plt.show()
```

```python
#Scatter Plot for Salary and Experience Level

plt.figure(figsize=(10, 6))
sns.scatterplot(x='experience_level', y='salary_in_usd', data=data, hue='experience_level')
plt.title('Salary in USD vs. Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()
```

```python
#salary distribution by job title

plt.figure(figsize=(12, 10))
sns.boxplot(x='job_title', y='salary_in_usd',palette='Accent', data=data.head(20))
plt.title('Salary Distribution by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary in USD')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#salary distribution over a time by company size.

plt.figure(figsize=(10, 6))
p = sns.lineplot(data=data, x='work_year', y='salary_in_usd', hue='company_size',ci=None, marker='o')

plt.xlabel('Year Work', fontsize=12, fontweight='bold')
plt.ylabel('Salary in USD', fontsize=12, fontweight='bold')

# Add a title
plt.title('Salary Trend Over Time by Company Size', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Group data by 'job_title' and calculate the average salary for each title

job_title_salary= data.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending = False)

plt.figure(figsize = (10,6))
p = sns.barplot(x= job_title_salary.values[:10], y = job_title_salary.index[:10],palette='Accent')

plt.title('Average Salary by Job Title (Top 10)', fontsize=12, fontweight='bold')
plt.xlabel('Average Salary (USD)', fontsize=12, fontweight='bold')
plt.ylabel('Job Title', fontsize=12, fontweight='bold')
               
plt.show()
```
